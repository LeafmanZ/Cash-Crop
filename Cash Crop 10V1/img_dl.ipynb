```python
# Import modules
import os, shutil
import pandas as pd
import urllib.request
import time
```

```python
def url_to_jpg(i, url, file_path):
    '''
        Args:
            -- i : number of image.
            -- url : a URL address of a given image.
            -- file_path : where to save the final image.
    '''

    filename = '-{}.jpg'.format(i)
    full_path = '{}{}'.format(file_path, filename)
    urllib.request.urlretrieve(url, full_path)

    print('{} saved.'.format(filename))

    return None
```

```python
list_csv = os.listdir(csv_dir)
print(list_csv)
```

```python
for csv in list_csv[2:]:
    name = csv[:-4]
    print('Beginning download for:', name)
    
    file_name = os.path.join(csv_dir, csv)
    file_path = os.path.join(img_dir, name)
    os.mkdir(file_path)
    final_file_path = os.path.join(file_path, name)
    urls = pd.read_csv(file_name)
    
    num_urls = len(urls.values)
    max_imgs = 8000
    sleep_btch_tm = 10
    sleep_csv_tm = 180
        
    if num_urls >= max_imgs:
        print('Downloading', max_imgs, 'images')
        for i, url in enumerate(urls.values[:max_imgs]):
            
            try:
                url_to_jpg(i, url[0], final_file_path)
            except:
                pass
            
            if i%250 == 0 and i!=0:
                print('DL paused for batch: Waiting', sleep_btch_tm ,'seconds....')
                time.sleep(sleep_btch_tm)
    
    else:
        print('Downloading', num_urls, 'images')
        for i, url in enumerate(urls.values):
            
            try:
                url_to_jpg(i, url[0], final_file_path)
            except:
                pass
            
            if i%250 == 0 and i!=0:
                print('DL paused for batch: Waiting', sleep_btch_tm ,'seconds....')
                time.sleep(sleep_btch_tm)

            
    print('DL paused for CSV: Waiting', sleep_csv_tm ,'seconds....')    
    time.sleep(sleep_csv_tm)
```
